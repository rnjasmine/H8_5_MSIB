## Combining Multiple Datasets
```
.concat()
```

````python
further_city_data = pd.DataFrame(
    {"revenue": [7000, 3400], "employee_count": [2, 2]},
    index=["New York", "Barcelona"]
)
````

````python
city_data
````

````python
further_city_data
````

````python
all_city_data = pd.concat([city_data, further_city_data], axis=0)
all_city_data
````

````python
all_city_data1 = pd.concat([city_data, further_city_data], sort=False)
all_city_data1
````

````python
city_countries = pd.DataFrame({
    "country": ["Holland", "Japan", "Holland", "Canada", "Spain"],
    "capital": [1, 1, 0, 0, 0]},
    index=["Amsterdam", "Tokyo", "Rotterdam", "Toronto", "Barcelona"]
)
city_countries
````

````python
cities = pd.concat([all_city_data, city_countries], axis=1)
cities
````

````python
cities1 = pd.concat([all_city_data, city_countries], axis=1, sort=False)
cities1
````

````python
cities.dropna()
````

````python
pd.concat([all_city_data, city_countries], axis=1, join="inner")
````

````python
countries = pd.DataFrame({
    "population_millions": [17, 127, 37],
    "continent": ["Europe", "Asia", "North America"]
}, index=["Holland", "Japan", "Canada"])
countries
````

````python
pd.merge(cities, countries, left_on="country", right_index=True)
````

````python
pd.merge(
    cities, 
    countries, 
    left_on="country", 
    right_index=True, 
    how="outer"
)
````
